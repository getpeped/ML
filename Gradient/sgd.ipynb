```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(44)

x = np.linspace(-2, 2, 100).reshape(-1, 1)
y = np.linspace(-3, 3, 100).reshape(-1, 1)

X = np.hstack((x, y, x**2, y**2))
Z = X[:, 2]**2 + X[:, 3]**2

# adding some Gaussian noise to the target value
# noise = np.random.normal(loc=0.0, scale=0.05, size=Z.shape) * Z
# Z = Z + noise

m = X.shape[0]
n_epochs = 50
t0, t1 = 5, 50

theta = np.random.randn(4, 1)
thetas = [theta.copy()] # learning step history

def learning_schedule(t):
    return t0 / (t + t1)

for epoch in range(n_epochs):
    for i in range(m):
        rand_idx = np.random.randint(m)
        xi = X[rand_idx: rand_idx + 1, :]
        yi = y[rand_idx: rand_idx + 1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients
        thetas.append(theta.copy())
                
th0 = theta[0, 0]
th1 = theta[1, 0]
th2 = theta[2, 0]
th3 = theta[3, 0]
print(f'The fuction reaches minumum for thetas: {th0:.4f}, {th1:.4f}, {th2:.4f}, {th3:.4f}')

XX, YY = np.meshgrid(X[:, 2], X[:, 3])
ZZ = th2 * XX + th3 * YY

# fig, ax = plt.subplots(figsize=(6, 6))
# ax.contour(XX, YY, ZZ, cmap='viridis', levels=20)


# thetas_arr = np.array(thetas).flatten().reshape(-1, 2) # ndarray [5001, 2]

```
